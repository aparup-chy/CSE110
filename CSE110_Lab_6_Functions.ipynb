```python
#task2

def fibonacchi(num):
    x = 0
    y = 1
    print(x, end=" ")
    print(y, end=" ")
    
    c = 0
    while c<num:
        new = x+y
        if new<num:
            print(new, end=" ")
            x = y
            y = new
        c +=1
fibonacchi(10)
```

```python
#7

def show_palindrome(val):
    a = ""
    for i in range(1,val+1):
        a += str(i) 
    return a+a[-2::-1]

x = show_palindrome(5)
print(x)
```

```python
#7 with space

def show_palindrome(val):
    a = ""
    for i in range(1,val+1):
        a += ' '+str(i) 
    return a[1:]+a[-2:0:-1]

x = show_palindrome(5)
print(x)
```

```python
#8

def show_palindrome(val):
    a = ""
    for i in range(1,val+1):
        a += ' '+str(i) 
    return a[1:]+a[-2:0:-1]

def show_palindromic_triangle(num):
    c = num-1
    x = ''
    for i in range(1,num+1):
        temp = " "*c
        c -= 1
        x += (2*temp+show_palindrome(i))+ '\n'
    return x
print(show_palindromic_triangle(5))
```

```python
#9

import math

def area_circumference_generator(rad):
    list1 = []
    circumference = 2*math.pi*rad
    area = math.pi*rad**2
    list1.append(circumference)
    list1.append(area)
    return tuple(list1)

print(area_circumference_generator(1))

x,y = area_circumference_generator(1)
print("Area of the circle is",x,"and circumference is",y)

```
